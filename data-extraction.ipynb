```python
import pandas as pd
from tqdm import tqdm
import json
import os

# Function to safely load JSON
def safe_load(f):
    try:
        return json.load(f)
    except json.JSONDecodeError:
        print(f"Error decoding JSON in file: {f.name}")
        return None

# Example list of filenames (replace with your actual list of file paths)
filenames = ['path/to/file1.json', 'path/to/file2.json', ...]  # Replace with actual filenames

# Initialize an empty list to collect all DataFrames
dfs = []

counter = 1
for file in tqdm(filenames):
    print(f"Processing file: {file}")  # Debug statement
    if os.path.exists(file):
        with open(file, 'r') as f:
            data = safe_load(f)
            if data is not None:
                df = pd.json_normalize(data)
                df['match_id'] = counter
                dfs.append(df)
                print(f"Appended DataFrame from file: {file}")  # Debug statement
                counter += 1
            else:
                print(f"Skipping file due to JSON decode error: {file}")
    else:
        print(f"File not found: {file}")

# Concatenate all DataFrames into a single DataFrame
if dfs:
    final_df = pd.concat(dfs, ignore_index=True)
else:
    final_df = pd.DataFrame()

print(final_df)  # Debug statement to check the final DataFrame
```
